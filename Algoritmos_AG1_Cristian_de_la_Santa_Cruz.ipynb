## Implementación de Mejoras

```python
# Complejidad O(n) - Método Iterativo
def Fibo_iter(n):
  f_0 = 0
  f_1 = 1
  if n<2:
    return n

  for i in range (2,n+1):
    f_2 = f_0 + f_1
    f_0 = f_1
    f_1 = f_2
  return f_2

Fibo_iter(9)
```

**Versión Divide y Vencerás**

Partiendo de la ecuación:

$$
\;
\left(\,
\begin{bmatrix}
0 & 1 \\
1 & 1
\end{bmatrix}
\right)^{n}
\;=\;
\begin{bmatrix}
f_{n-1} & f_{n} \\
f_{n} & f_{n+1}
\end{bmatrix}
$$

Realizamos la técnica de divide y vencerás convirtiendo la matriz n-ésima en productos de diferentes matrices de potencias de orden 1 y orden 2.

$$
x^{n} =
\begin{cases}
x,        & n = 1,\\
(x^{n/2})^2,              & n \text{ es par},\\
x*x^{n-1},     & n \text{ es impar}.
\end{cases}
$$

```python
# Divide y Vencerás Fibonacci
# Complejidad O(log(n))
def Fibo_DV(n):
  if n <= 0:
    return 0
  i = n-1
  f0, f1 = 0, 1
  a, b, c, d = f1, f0, f0, f1
  while i > 0:
    if i % 2 == 1: # es impar
      f0 = d*b + c*a
      f1 = d*(b+a)+c*b
      a, b = f0, f1
    f0 = c**2 + d**2
    f1 = d*(2*c+d)
    c,d = f0,f1
    i=i//2
  return a+b

Fibo_DV(9)
```

**Fórmula de Binet**

$$
F_n = \frac{\varphi^n - \psi^n}{\sqrt{5}}
$$

$$
\varphi = \frac{1 + \sqrt{5}}{2},
\quad
\psi = \frac{1 - \sqrt{5}}{2}.
$$

```python
# Resolución a través de fórmula de Binet
# Complejidad O(1)
from math import sqrt
def Fibo_Binet(n):
  if n<2:
    return n
  else:
    fi = (1 + sqrt(5)) / 2 # Proporción Aurea (Autovalores)
    Fn = (fi**n - (1 - fi)**n) / sqrt(5)
    return round(Fn)

Fibo_Binet(9)
```

## Mejora: Metodos de Ordenación de Listas (caso en que SM no esté ordenado)

```python
def insercion(SM: list):
    for i in range(1, len(SM)):
        valor = SM[i]
        j = i - 1
        # Mover todos los elementos mayores que valor hacia adelante
        while j >= 0 and valor > SM[j]:
            j -= 1
        # Eliminar el valor original e insertarlo en la nueva posición
        SM.pop(i)
        SM.insert(j + 1, valor)
    return SM

print(insercion([34, 45, 44, 47]))
```

```python
def burbuja(SM: list): #O(n) en el mejor caso O(n^2) en el peor caso
  "Recorre la lista para intercambiar cada elemento con su adyacente si es preciso en cada iteracción"
  # cambios = Variable que almacena si hay cambios a lo largo de una pasada
  for i in range(len(SM)):
    cambios = False
    for j in range(len(SM)-i-1): # j se actualiza y va desde el elemento i + 1
                                 # hasta len(SM) puesto que los elementos antes de i ya están ordenados
      if j + 1 < len(SM) and (SM[j]>SM[j+1]):
        valor = SM[j+1] # guardo el valor i+1
        SM[j+1] = SM[j] # paso el valor i a i+1
        SM[j] = valor # devuelvo el valor i+1 a i
        cambios = True
    if not cambios:
      break
  return SM

burbuja([4,35,23,16,44,55,5,242,3,45,564,6456432,61635,8035,345,36247826])
```

```python
def seleccion (SM: list): # O(n^2) Algoritmo de selección con mejora sitúa el máximo y el mínimo de una sola pasada
  max, min = SM[0], SM[0]
  iteracion = 0
  # Obtener la posición con el valor minimo
  while len(SM) - 2*iteracion >= 0:
    min_posicion = iteracion
    max_posicion = iteracion
    max, min = SM[iteracion], SM[iteracion]

    for i in range(len(SM) - 2*iteracion):
      if min > SM[i+iteracion]:
        min_posicion = i+iteracion
        min = SM[i+iteracion]
      if max < SM[i+iteracion]:
        max_posicion = i+iteracion
        max = SM[i+iteracion]

    # Intercambiar valor
    valor_fin = SM[-1-iteracion]
    SM[-1-iteracion] = SM[max_posicion]
    SM[max_posicion] = valor_fin

    if min_posicion == len(SM) - 1 - iteracion:
      min_posicion = max_posicion

    valor_inicio = SM[iteracion]
    SM[iteracion] = SM[min_posicion]
    SM[min_posicion] = valor_inicio

    iteracion += 1

  return SM

seleccion([4,35,23,16,44,55,43,5,242,3,45,564,55555555555,6456432,61635,8035,345,36247826])
```

# N-Reinas por técnica de vuelta atrás

```python
def escribe(S):
  n = len (S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X ", end="")
      else:
        print(" - ", end="")

def es_prometedora(SOLUCION, etapa):
  #print(SOLUCION)
  #SI la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor" + str(SOLUCION[i])) + "está" + str(SOLUCION.count(SOLUCION[i])) + "veces")
    if SOLUCION.count(SOLUCION[i]) > 1: # Contar si dentro de la solución aparece más de un valor igual
      return False

    # Verifica las diagonalesç
    for j in range(i+1, etapa+1):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
    solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        reinas(N,solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

# Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
